### 1.3.1.Descripción:

	- instant: Indice
	- dteday : Dia
	- season : Temporada (1:Primavera, 2:Verano, 3:Otoño, 4:Invierno)
	- yr : Año (0: 2011, 1:2012)
	- mnth : Mes ( 1 to 12)
	- hr : hora (0 to 23)
	- holiday : el dia es feriado o no (extracted from http://dchr.dc.gov/page/holiday-schedule)

	- workingday : si el día no es fin de semana o festivo es 1, sino es 0.

		- 2: Niebla + nubes
		- 3: Nieve ligera, lluvias ligeras

	- temp : Temperatura en Celsius. 
	- atemp: Sensación termica en Celsius.
	- hum: porcentaje de humedad
	- windspeed: Velocidad del viento
	- casual: Usuarios casuales
	- registered: usuarios registrados
	- cnt: total de usuarios (casual + registered)

-Veggies:  Consume Vegetables 1 or more times per day 0 = no 1 = yes

-NoDocbcCost:  Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? 0 = no 1 = yes

-DiffWalk:  Do you have serious difficulty walking or climbing -stairs? 0 = no 1 = yes
-Sex:  0 = female 1 = male
-Ages:  13-level age category (_AGEG5YR see codebook) 1 = 18-24 9 = 60-64 13 = 80 or older
-Education:  Education level (EDUCA see codebook) scale 1-6 1 = -Never attended school or only kindergarten 2 = Grades 1 through 8 (Elementary) 3 = Grades 9 through 11 (Some high school) 4 = Grade 12 or GED (High school graduate) 5 = College 1 year to 3 years (Some college or technical school) 6 = College 4 years or more (College graduate)
-Income:  Income scale (INCOME2 see codebook) scale 1-8 1 = less than 10,000 5 = less than 35,000 8 = 75,000 or more
