```python
import requests



url = "http://api.citybik.es/v2/networks/"

response = requests.request("GET", url)

print(response.text)


```

```python
import json
import pandas as pd
datos = response.json()
```

```python
print(datos)
```

```python
res = pd.DataFrame(datos)
```

```python
res
```

```python



df_limpio = pd.json_normalize(datos,record_path="networks")
df_limpio
```

```python
import folium
from folium.plugins import MiniMap
bicis = folium.Map()
```

```python


bicis = folium.Map  (location=[df_limpio.iloc[0]["location.latitude"],df_limpio.iloc[0]["location.longitude"]] , zoom_start = 4)
for a in range(0,615):
  folium.Marker (location=[df_limpio.iloc[a]["location.latitude"],df_limpio.iloc[a]["location.longitude"]], popup = df_limpio.iloc[a]["name"]).add_to(bicis)
  





bicis
```
